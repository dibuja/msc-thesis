```python
import rpy2.robjects as robjects
from rpy2.robjects import pandas2ri
pandas2ri.activate()

readRDS = robjects.r['readRDS']

# Source file: https://dataverse.harvard.edu/file.xhtml?persistentId=doi:10.7910/DVN/L4OAKN/PGOQ6T&version=1.0
df = readRDS('data/Corp_Congreso_V2.rds')
df = pandas2ri.rpy2py_dataframe(df)
```

The corpora found in Harvard dataverse is not useful for me since:

- It excludes the political group speakers belong to.
- It includes procedural text as part of speaker interventions.
- It excludes the topic that is being discussed.
- It excludes the legislature that the speech belongs to.

Therefore, I decided to ignore it and build the dataset myself pulling the data from the https://www.congreso.es database and by performing PDF scraping.
